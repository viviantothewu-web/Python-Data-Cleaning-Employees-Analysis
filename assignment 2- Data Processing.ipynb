**Task 1: NdArray (5 Points)**

- Create a list and 1-dimensional numpy array with identical elements and show that slicing in both is the same.
- Choose 3 movies and create 2, 3x4 numpy arrays as follows:
- The first numpy array called cast contains the name of 4 actors/actresses from each movie.
- The second numpy array called movies contains the name of the movie corresponding to each cast.
- Calculate the boolean indexing for one of the movies in the second numpy array.
- Using the boolean indexing from the previous step, select the actors/actresses that played in that movie and put it in a new numpy array called selected_movie.
- Save all of the three numpy arrays from the previous steps to disk.
- Reload the saved file and load the three numpy arrays to three new variables called cast_new, movies_new and selected_movie_new.

```python
# Create a list and 1-dimensional numpy array with identical elements and show that slicing in both is the same.
import numpy as np

sample_list = ['apples', 'bananas', 'orange', 'grapes']
sample_np = np.array(['apples', 'bananas', 'orange', 'grapes'])

#slicing
print (sample_list[0:3])
print (sample_np[0:3])
```

**Explanation**
results showed that list and numpy array slice the same way

```python
# Choose 3 movies and create 2, 3x4 numpy arrays 
# first numpy array called cast
cast = np.array([
    ['Daniel Radcliff', 'Emma Watson', 'Rupert Grint', 'Tom Felton'],
    ['Leonardo Dicaprio', 'Kate Winslet', 'Billy Zane', 'Kathy Bates'], 
    ['Matt Damon', 'Jessica Chastain', 'Kristen Wiig', 'Jeff Daniels']
])
# second numpy array called movies
movies = np.array ([
    ['Harry Potter', 'Harry Potter', 'Harry Potter', 'Harry Potter'],
    ['Titanic', 'Titanic', 'Titanic', 'Titanic'],
    ['Martian', 'Martian', 'Martian', 'Martian']
])
```

**Explanation**
Made 2 numpy arrays with same shapes. One with 4 cast names from each movie, and one with just the movie

```python
# Calculate the boolean indexing for one of the movies in the second numpy array
boolean_index = movies == 'Titanic'
print (boolean_index)
```

**Explanation**
One example of boolean indexing I used is to see which row matches 'Titanic'. As shown in the result, you can see the second row matches.

```python
# Using the boolean indexing from the previous step, select the actors/actresses that played in that movie and put it in a new numpy array called selected_movie.
selected_movie = cast[boolean_index]
print (selected_movie)
```

**Explanation**
I selected the cast that's True to the boolean_index, which are casts of Titanic.

```python
# Save all of the three numpy arrays from the previous steps to disk.
np.savez(
    'movie_data.npz',
    cast=cast,
    movies=movies,
    selected_movie=selected_movie
)
print ("all numpy arrays saved")
```

**explanation**
save to a numpy zip file, where the keys (arrays) are cast, movies, and selected movie

```python
#Reload the saved file and load the three numpy arrays to three new variables
#called cast_new, movies_new and selected_movie_new.
data = np.load('movie_data.npz')
cast_new = data['cast']
movies_new = data['movies']
selected_movie_new = data['selected_movie']
```

**explanation**
loaded the file again and gave new names

```python
print(cast_new)
```

```python
print(movies_new)
```

```python
print(selected_movie_new)
```

printed all new names to check

**Task 3: DataFrame (7 Points)**

Create a DataFrame as follows:

- The DataFrame columns are "fruit" (string), "number_of_seeds" (integer), "edible" (boolean) and "season" (one of the values for 'summer', 'fall' or 'spring'), and the DataFrame contains at least 5 rows.
- Add a new column to the DataFrame called "Color" and show that all the values for the column are NaN after adding the column
- Set all of the values for the "Color" column to "Unknown"
- Use slicing properly to extract "number_of_seeds" and "fruit" for 2 rows.
- Write a function called "seed_count_calculator" and use the apply() method to sum the number_of_seeds row. (do not use the built-in sum functions for this part)
- Try to sort the DataFrame by fruit names.
- Find both the index and name of the fruit with the maximum amount of seeds.
- Save the DataFrame to disk as a CSV file.
- Reload the saved CSV in a new variable.

```python
import pandas as pd

data = {
    'fruit': ['apple', 'banana', 'orange', 'grape', 'pineapple'],
    'number_of_seeds':[5, 0, 3, 1, 4],
    'edible': [True, True, True, True, True],
    'season': [ 'Summer', 'Summer', 'Winter', 'Summer', 'Summer']
}

df = pd.DataFrame(data)
df
```

**explanation**
made a table based on the instruction, need at least 5 rows so added 5 things to each category (column)

```python
#Add a new column to the DataFrame called "Color" and show that all the values for the column are NaN after adding the column
df['color'] = None
df
```

**explanation**

Tried doing df['color'] = Nan, but it says its not defined, so I changed it to None and showed the dataframe

```python
# Set all of the values for the "Color" column to "Unknown"
df['color'] = 'Unknown'
df
```

```python
# Use slicing properly to extract "number_of_seeds" and "fruit" for 2 rows.
df.loc[0:1, ['number_of_seeds', 'fruit']]
```

**explanation**
slicing: [0:1] gets the first 2 rows
.loc is label based and I looked up number of seeds and fruits

```python
# Write a function called "seed_count_calculator" and use the apply() method to sum the number_of_seeds row. 

def seed_count_calculator(series):
    total = 0
    for value in series:
        total += value
    return total

total_seeds = seed_count_calculator(df['number_of_seeds'].apply(lambda x: x))
print(total_seeds)
```

**explanation**
df[number of seeds] is a series which i created a function for. I started total count as 0, and every time it loops through the number in series, it adds to total.
apply() runs the function through each value and lambda doesn't do anything in this case, it just returns the value and print total seeds

```python
# Try to sort the DataFrame by fruit names.
df.sort_values(by= 'fruit')
```

**explanation**
sorted fruit by alphabetical order

```python
#Find both the index and name of the fruit with the maximum amount of seeds.
df.loc[df['number_of_seeds'].idxmax(), 'fruit']
```

**explanation**
number of seeds.idxmax finds the max amount of seeds and .loc (label based as mentioned) find the name of the fruit

```python
# Save the DataFrame to disk as a CSV file.
df.to_csv('Fruitsdf.csv', index = False)
```
