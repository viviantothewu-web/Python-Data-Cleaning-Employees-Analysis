```python
# assigning buildings for each team

mapping = {
    #building 1
    "Marketing": "Building1",
    "Client Services": "Building1",
    "Product": "Building1", 
    "Sales": "Building1",

    # building 2
    "Finance": "Building2", 
    "Legal": "Building2" , 
    "Human Resources": "Building2", 

    #building3
    "Engineering": "Building3", 
    "Business Development": "Building3",

    #building4
    "Distribution": "Building4"
}
```

**explanation**
I assigned building numbers to each team based on the instruction

```python
# create a new column in csv_df called location and fill its values according to the value of Team column
csv_df["Location"] = csv_df["Team"].map(mapping).fillna("Administration")
csv_df.head()
```

**explanation**

i used map to apply the building numbers to each team, and fill na with "Administration". at llast I assigned the building numbers to a new column name location. As the results shown, the last column is a new column I added with the building number corresponding to the teams

```python
# Calculate the average of salary and bonus based on the Team

average_salary= csv_df.groupby('Team')[['Salary','Bonus %']].mean()
average_salary
```

**explanation**
using groupby, I grouped csv_df by "Teams" and then only looked at the "Salary" and "Bonus" columns and calculated the mean. The result showed the average salaries for each team

```python
# Create a function called top() that takes a DataFrame as an input and returns the 5 highest paid entries in the DataFrame
def top (dataframe):
    return dataframe.sort_values('Salary', ascending = False).head(5)

# Then use the top() function to get the 5 highest paid people of each Gender. 
Highest_paid = (
    csv_df
    .groupby("Gender")
    .apply(top)
)

Highest_paid
```

**explanation**
created a funciton name top first that returns top 5 paid employees using sort values then getting the first 5 rows.
then I groupedby gender and applied the top function. The result showed highest paid top 5 employees in each gender group.
